```python
import numpy as np

while True:
  try:
      user_input = input("Porfavor ingrese radios (números positivos) separados por comas.")
      radios = [int(num.strip()) for num in user_input.split(",")]
      break
  except ValueError:
      print("Por favor ingrese solo números positivos separados por comas.")

perimetros = []
areas = []
for r in radios :
      area = np.pi * r**2
      perimetro = 2 * np.pi * r
      perimetros.append(perimetro)
      areas.append(area)
print("Los perímetros y áreas de los círculos respectivos a los radios ingresados son:")
print("Perímetros:", perimetros)
print("Áreas:", areas)
```
